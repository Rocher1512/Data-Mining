# Cleaning

## General

```python
#Imports
import pandas as pd
import time as time
import seaborn as sns
import matplotlib.pyplot as plt
import scikitplot as skplt
import numpy as np

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import plot_tree, DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer

#Global lists
alredythere = []
listofwrongs = []
listofcolumns = []
listfixed = []

#Searching for any incorect records
def Creatingthelist(srs_a):
    lenth = srs_a.size
    for x in range(lenth):
        if pd.isna(srs_a[x]) == True or srs_a[x] == "Unknown" or srs_a[x] == ' ':
            if x in alredythere:
                x
            else:
                alredythere.append(x)
                listofwrongs.append(Data.loc[x])

#Cleaning Indication so everything is capitalised 
def CleaningIndication(srs_a):
    lenth = srs_a.size
    for x in range(lenth):
        if(srs_a[x] == 'CVA' or srs_a[x] == 'A-F' or srs_a[x] == 'ASX' or srs_a[x] == 'TIA'):
            x
        else:
            FreshData.at[x, 'Indication'] = str(srs_a[x]).upper()

#loading in the data
Data = pd.read_csv("2020-DMDS-ACWData_fixed.csv")
FreshData = Data
listofcolumns = FreshData.columns

#cycling through every column

for x in range(len(listofcolumns)):
    srs_a = FreshData[listofcolumns[x]]
    Creatingthelist(srs_a)
CleaningIndication(FreshData['Indication'])
```

## Deletion 

```python
#Timing using https://docs.python.org/3/library/time.html
ST = time.perf_counter()
CleanData = FreshData
for x in range(len(alredythere)):
     CleanData = CleanData.drop(alredythere[x])
ET = time.perf_counter()    
Timetaken = ET - ST
print(Timetaken)
```

## Mode Mean

```python
#general calulation of mode
def mode(srs_a):
    lenth = srs_a.size
    alreadyin = []
    count = []
    place = 0
    mode = 0
    modenum = 0
    for x in range(lenth):
        for j in range(len(alreadyin)):
            if(alreadyin[j] == srs_a[x]):
                count[j] = count[j] +1
                place = [j]
                break
        if(place == 0):
            count.append(1)
            alreadyin.append(srs_a[x])
    for x in range(len(count)):
        if(len(count) > modenum):
            modenum = len(count)
            mode = alreadyin[x]
    return mode

#general calulation of mean
def mean(srs_a):
    lenth = srs_a.size
    count = 0
    for x in range(lenth):
        if pd.isna(srs_a[x]) == False:
            if srs_a[x] != ' ':
                count = count + int(srs_a[x])
    count = count/lenth
    return count

ST = time.perf_counter()
CleanData = FreshData
#checking each column for incorect values 
Indicationmode = mode(CleanData['Indication'])
Diabetesmode = mode(CleanData['Diabetes'])
IHDmode = mode(CleanData['IHD'])
Hypertensionmode = mode(CleanData['Hypertension'])
Arrhythmiamode = mode(CleanData['Arrhythmia'])
Historymode = mode(CleanData['History'])
IPSImean = mean(CleanData['IPSI'])
Contramean = mean(CleanData['Contra']) 
labelmode = mode(CleanData['label'])
IPSImean = int(IPSImean)
Contramean = int(Contramean)

#replacing incorect values 
for x in range(len(listofwrongs)):
    temp = listofwrongs[x]
    NAvalue = ''
    for j in range(len(listofcolumns)):
        if pd.isna(temp.loc[listofcolumns[j]]) == True or temp.loc[listofcolumns[j]] == "Unknown" or temp.loc[listofcolumns[j]] == ' ':
            NAvalue = listofcolumns[j]
    if NAvalue == 'Indication':
        CleanData.at[alredythere[x], NAvalue] = Indicationmode
    elif NAvalue == 'Diabetes':
        CleanData.at[alredythere[x], NAvalue] = Diabetesmode
    elif NAvalue == 'IHD':
        CleanData.at[alredythere[x], NAvalue] = IHDmode
    elif NAvalue == 'Hypertension':
        CleanData.at[alredythere[x], NAvalue] = Hypertensionmode
    elif NAvalue == 'Arrhythmia':
        CleanData.at[alredythere[x], NAvalue] = Arrhythmiamode
    elif NAvalue == 'History':
        CleanData.at[alredythere[x], NAvalue] = Historymode
    elif NAvalue == 'IPSI':
        CleanData.at[alredythere[x], NAvalue] = IPSImean
    elif NAvalue == 'Contra':
        CleanData.at[alredythere[x], NAvalue] = Contramean
    elif NAvalue == 'label':
        CleanData.at[alredythere[x], NAvalue] = labelmode

ET = time.perf_counter()    
Timetaken = ET - ST
print(Timetaken)
```

## Nearest Neighbor 

```python
ST = time.perf_counter()
CleanData = FreshData
for x in range(len(listofwrongs)):
    temp = listofwrongs[x]
    NAvalue = ''
    for j in range(len(listofcolumns)):
        if pd.isna(temp.loc[listofcolumns[j]]) == True or temp.loc[listofcolumns[j]] == "Unknown" or temp.loc[listofcolumns[j]] == ' ':
            NAvalue = listofcolumns[j]
    largestclosest = 0
    closestdata = CleanData.loc[0]
    for j in range(len(CleanData)):
        closest = 0
        testing = CleanData.loc[j]
        if(testing['Id'] == temp['Id']):
            continue
        if(testing['Indication'] == temp['Indication']):
            closest = closest + 1
        if(testing['Diabetes'] == temp['Diabetes']):
            closest = closest + 1
        if(testing['IHD'] == temp['IHD']):
            closest = closest + 1
        if(testing['Hypertension'] == temp['Hypertension']):
            closest = closest + 1
        if(testing['Arrhythmia'] == temp['Arrhythmia']):
            closest = closest + 1
        if(testing['History'] == temp['History']):
            closest = closest + 1
        try:
            if((testing['IPSI'] > (temp['IPSI'] - 10)) & (testing['IPSI'] < (temp['IPSI'] + 10))):
                closest = closest + 1
        except:
            closest
        try:
            if((int(testing['Contra']) > (int(temp['Contra']) - 10)) & (int(testing['Contra'])  < (int(temp['Contra']) + 10))):
                closest = closest + 1
        except:
            closest
        if(testing['label'] == temp['label']):
            closest = closest + 1
        if(closest > largestclosest):
            largestclosest = closest
            closestdata = testing
    CleanData.at[alredythere[x], NAvalue] = closestdata[NAvalue]

ET = time.perf_counter()    
Timetaken = ET - ST
print(Timetaken)
```

# Model

## Multilayer Perceptron

```python
def MLP(x_train,y_train,x_test):
    ST = time.perf_counter()
    model = MLPClassifier(solver='sgd', alpha = 0.0001, hidden_layer_sizes=(370,370,370), activation = 'tanh', learning_rate = 'invscaling')
    trained_model = model.fit(x_train, y_train)
    predict = trained_model.predict(x_test)
    ET = time.perf_counter()    
    Timetaken = ET - ST
    print(Timetaken)
    return predict
```

## DecisionTreeClassifier

```python
def DTC(x_train,y_train,x_test):
    ST = time.perf_counter()
    model = DecisionTreeClassifier(criterion = 'gini', max_depth = 12, splitter = 'best') 
    trained_model = model.fit(
        x_train,
        y_train)

    predict = trained_model.predict(x_test)
    ET = time.perf_counter()    
    Timetaken = ET - ST
    print(Timetaken)
    return predict
```

## Random  Forest Classifier

```python
def RFC(x_train,y_train,x_test):    
    ST = time.perf_counter()
    model = RandomForestClassifier(criterion = 'entropy', max_depth = 13, n_estimators = 50)
    trained_model = model.fit(
        x_train,
        y_train)

    predict = trained_model.predict(x_test)
    ET = time.perf_counter()    
    Timetaken = ET - ST
    print(Timetaken)
    return predict
```

## Extra Tree Classifier 

```python
def ETC(x_train,y_train,x_test):    
    ST = time.perf_counter()
    model = ExtraTreesClassifier(criterion = 'entropy', max_depth = 13, n_estimators = 50)
    trained_model = model.fit(
        x_train,
        y_train)

    predict = trained_model.predict(x_test)
    ET = time.perf_counter()    
    Timetaken = ET - ST
    print(Timetaken)
    return predict
```

## Gaussian Naive Bayes

```python
def GNB(x_train,y_train,x_test):
    ST = time.perf_counter()
    model = GaussianNB()
    trained_model = model.fit(
        x_train,
        y_train)

    predict = trained_model.predict(x_test)
    ET = time.perf_counter()    
    Timetaken = ET - ST
    print(Timetaken)
    return predict
```

## Voting Classifier

```python
def VC(x_train,y_train,x_test):
    ST = time.perf_counter()
    clf1 =  MLPClassifier(solver='sgd', alpha = 0.0001, hidden_layer_sizes=(370,370,370), activation = 'tanh', learning_rate = 'invscaling')
    clf2 = RandomForestClassifier(criterion = 'entropy', max_depth = 13, n_estimators = 50)
    clf3 = GaussianNB()
    model = VotingClassifier(estimators=[('MLP', clf1), ('RFC', clf2), ('GNB', clf3)], voting='hard')

    trained_model = model.fit(
        x_train,
        y_train)
    predict = trained_model.predict(x_test)
    ET = time.perf_counter()    
    Timetaken = ET - ST
    print(Timetaken)
    return predict
```

### K Means

```python
#Spliting the data 5 ways
X = New_data.loc[:, New_data.columns != 'label_Risk']
Y = New_data['label_Risk']
X_1 = X 
X_2 = X
X_3 = X
X_4 = X
X_5 = X
Y_1 = Y
Y_2 = Y
Y_3 = Y
Y_4 = Y
Y_5 = Y
for i in range(len(X)):
    if i < 304:
        X_2 = X_2.drop([i])
        Y_2 = Y_2.drop([i])
        X_3 = X_3.drop([i])
        Y_3 = Y_3.drop([i])
        X_4 = X_4.drop([i])
        Y_4 = Y_4.drop([i])
        X_5 = X_5.drop([i])
        Y_5 = Y_5.drop([i])
    elif i < 608:
        X_1 = X_1.drop([i])
        Y_1 = Y_1.drop([i])
        X_3 = X_3.drop([i])
        Y_3 = Y_3.drop([i])
        X_4 = X_4.drop([i])
        Y_4 = Y_4.drop([i])
        X_5 = X_5.drop([i])
        Y_5 = Y_5.drop([i])
    elif i < 911:
        X_1 = X_1.drop([i])
        Y_1 = Y_1.drop([i])
        X_2 = X_2.drop([i])
        Y_2 = Y_2.drop([i])
        X_4 = X_4.drop([i])
        Y_4 = Y_4.drop([i])
        X_5 = X_5.drop([i])
        Y_5 = Y_5.drop([i])
    elif i < 1215:
        X_1 = X_1.drop([i])
        Y_1 = Y_1.drop([i])
        X_2 = X_2.drop([i])
        Y_2 = Y_2.drop([i])
        X_3 = X_3.drop([i])
        Y_3 = Y_3.drop([i])
        X_5 = X_5.drop([i])
        Y_5 = Y_5.drop([i])
    else:
        X_1 = X_1.drop([i])
        Y_1 = Y_1.drop([i])
        X_2 = X_2.drop([i])
        Y_2 = Y_2.drop([i])
        X_3 = X_3.drop([i])
        Y_3 = Y_3.drop([i])
        X_4 = X_4.drop([i])
        Y_4 = Y_4.drop([i])
        
def K_mean (model):
    X_T = [X_1,X_2,X_3,X_4,X_5]
    Y_T= [Y_1,Y_2,Y_3,Y_4,Y_5]
    P_Y = []
    Predict = []
    for i in range(5):
        #creating the training and test data sets
        temp_X = X_T
        temp_Y = Y_T
        temp_X.pop(i)
        temp_Y.pop(i)
        train_x = pd.concat(temp_X[j] for j in range(len(temp_X)))                 
        train_y = pd.concat(temp_Y[j] for j in range(len(temp_Y)))
        X_T = [X_1,X_2,X_3,X_4,X_5]
        Y_T= [Y_1,Y_2,Y_3,Y_4,Y_5]
        #traing the model
        temp = model(train_x,train_y,X_T[i])
        P_Y.append(temp)
    for i in range(len(P_Y)):
        for j in range(len(P_Y[i])):
            Predict.append(P_Y[i][j])
    real_Y = pd.concat(Y_T[j] for j in range(len(Y_T)))
    #confusion matrix
    skplt.metrics.confusion_matrix(Predict, real_Y)
    skplt.metrics.plot_confusion_matrix(
        Predict,
        real_Y,
        normalize = False
    )
    skplt.metrics.plot_confusion_matrix(
        Predict,
        real_Y,
        normalize = True
    )
print("MLP")
K_mean(MLP)
print("DTC")
K_mean(DTC)
print("RFC")
K_mean(RFC)
print("ETC")
K_mean(ETC)
print("GNB")
K_mean(GNB)
print("VC")
K_mean(VC)
```

### Grid Search

```python
#Grid Search
#Code used from https://towardsdatascience.com/hyperparameter-optimization-with-scikit-learn-scikit-opt-and-keras-f13367f3e796 
ST = time.perf_counter()
x_train,x_test,y_train,y_test  = train_test_split(New_data.loc[:, New_data.columns != 'label_Risk'], New_data['label_Risk'], test_size=0.3, shuffle=True)

param_grid = {'bootstrap': [True],
     'max_depth': [6, 10],
     'max_features': ['auto', 'sqrt'],
     'min_samples_leaf': [3, 5],
     'min_samples_split': [4, 6],
     'n_estimators': [100, 350]
    }
     

forest_clf = RandomForestClassifier()

forest_grid_search = GridSearchCV(forest_clf, param_grid, cv=5,
                                  scoring="accuracy",
                                  return_train_score=True,
                                  verbose=True,
                                  n_jobs=-1)

forest_grid_search.fit(x_train, y_train)
print(forest_grid_search.best_score_)
ET = time.perf_counter()    
Timetaken = ET - ST
print(Timetaken)
```

```python
# MLP custom

x_train,x_test,y_train,y_test  = train_test_split(New_data.loc[:, New_data.columns != 'label_Risk'], New_data['label_Risk'], test_size=0.3, shuffle=True)
target = False

def modeltrain(ly1,ly2,ly3):
    model1 = MLPClassifier(solver='sgd', alpha = 0.0001, hidden_layer_sizes=(ly1,ly2,ly3), activation = 'tanh', learning_rate = 'invscaling')
    model2 = MLPClassifier(solver='sgd', alpha = 0.0001, hidden_layer_sizes=(ly1,ly2,ly3), activation = 'tanh', learning_rate = 'invscaling')
    model3 = MLPClassifier(solver='sgd', alpha = 0.0001, hidden_layer_sizes=(ly1,ly2,ly3), activation = 'tanh', learning_rate = 'invscaling')
    model1.fit(x_train,y_train)
    model2.fit(x_train,y_train)
    model3.fit(x_train,y_train)
    print(model1.score(x_test,y_test))
    print(model2.score(x_test,y_test))
    print(model3.score(x_test,y_test))
    score = (model1.score(x_test,y_test) + model2.score(x_test,y_test) + model3.score(x_test,y_test))/3
    return score
bestscore = 0
bestweights = [0,0,0]
for x in range(1,501,10):
    for j in range(1,501,10):
        for k in range(1,501,10):
            if(target == False):
                score = modeltrain(x,j,k)
                if(score > bestscore):
                    bestscore = score
                    bestweights[0]=x
                    bestweights[1]=j
                    bestweights[2]=k
                    
print("Best weights: ")
for i in range(3):
    print(bestweights[i])
    
```
